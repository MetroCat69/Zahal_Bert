```python
def read_pdf(path):
    all_text = ''
    reader = PdfReader(path)
    number_of_pages = len(reader.pages)
    for i in range(0,number_of_pages):
        page = reader.pages[i]
        text = page.extract_text()
        all_text+=text+"<next_page>"
        
    return all_text
```

```python
def keep_hebrew_and_ascii(text):
    filtered_text = ''
    for char in text:
        # Check if the character is Hebrew, ASCII, newline, or tab
        if ('\u0590' <= char <= '\u05FF') or ('\u0020' <= char <= '\u007F') or char in ['\n', '\t']:
            filtered_text += char
    return filtered_text

# Example usage
text = "שלום, Hello! 你好！\n\tTesting.{}\\..////;;;;."
filtered_text = keep_hebrew_and_ascii(text)
print(filtered_text)
```
