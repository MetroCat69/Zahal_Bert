```python
def find_unique_items(list1, list2):
    # Find items that are only in the first list
    unique_to_list1 = set(list1) - set(list2)
    
    # Find items that are only in the second list
    unique_to_list2 = set(list2) - set(list1)
    
    return unique_to_list1, unique_to_list2
```

```python
unique_vocab_new, unique_vocab_old = find_unique_items(new_vocab, old_vocab)
```

```python
zahal_corpus_df['context'].iloc[0:2].map(lambda x : x.replace('"', "")).map(lambda x : x.replace('-', "")).map(lambda x : x.replace('.', " ")).map(lambda x : x.replace(',', " "))
```

```python
tokenized_text = zahal_corpus_df['context'].map(lambda x : x.replace('"', "")).map(lambda x : x.replace('-', "")).map(lambda x : x.replace('.', " ")).map(lambda x : x.replace(',', " ")).map(lambda x : x.replace(';', " ")).map(lambda x : x.replace(':', " ")).map(lambda x : x.replace('(', " ")).map(lambda x : x.replace(')', " ")).map(lambda x : x.replace("'", " ")).map(lambda x : x.replace("!", " ")).map(lambda x : x.replace("?", " ")).map(lambda x : x.replace("\n", " ")).map(lambda x : x.replace("\t", " ")).map(lambda x : x.replace("\\", " ")).map(lambda x : x.replace("/", " ")).apply(word_tokenize)
```

```python
freq_dist = FreqDist([word for sublist in tokenized_text for word in sublist])
```

```python
most_common_words = freq_dist.most_common(1000000)
```

```python
zahal_words_df = pd.DataFrame({'names' : pd.Series(most_common_words).map(lambda x: x[0]),'n_ocurrences' : pd.Series(most_common_words).map(lambda x: x[1])})
```

```python
pd.set_option('display.max_rows', 200)
new_vocab_to_add = zahal_words_df.loc[(zahal_words_df['names'].map(lambda x : x not in old_vocab))&(zahal_words_df['names'].map(lambda x : len(x) )>2)][0:10000]
```

```python
pd.DataFrame({'vocab':new_vocab_to_add['names']}).to_csv('vocab.csv')
```
