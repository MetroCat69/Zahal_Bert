```python
import re
# Split the string based on multiple sequences
sequences = ["\n ", "\t", "<next_page>","\n1."]

# Construct the regular expression pattern
pattern = "|".join(re.escape(seq) for seq in sequences)

# Split the string using the pattern
text_list = re.split(pattern, x)
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')
```

```python
text_df = pd.DataFrame({'text' : text_list})
```

```python
text_df["len"] = text_df["text"].map(lambda x :len(x))
```

```python
text_df_filtered = text_df.loc[text_df["len"] > 5]
```

```python
def get_embeddings(text_list):
   
    return  model.encode(text_list)
```

```python
text_df_filtered["text"]
```

```python
text_df_filtered['embed'] = text_df_filtered['text'].map(lambda x :get_embeddings(x))
```
